#Data preprocessing

````python
import numpy as np
import pandas as pd
import tensorflow as tf
````

````python
tf.__version__
````



```
# LINK FOR DATASET: https://archive.ics.uci.edu/dataset/294/combined+cycle+power+plant
```



````python
data = pd.read_excel('/content/data.xlsx')
x = data.iloc[:, :-1].values
y = data.iloc[:,-1].values
````

````python
print('X = ',x)
print('Y = ',y)
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
````

````python
print('X_train',X_train)
print('Y_train',Y_train)
print('X_test',X_test)
print('Y_test',Y_test)
````

#Building ANN

````python
ann = tf.keras.models.Sequential()
````

````python
ann.add(tf.keras.layers.Dense(units = 6,activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 6,activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 1))
````

````python
ann.compile(optimizer = 'adam', loss = 'mean_squared_error')
````

````python
ann.fit(X_train,Y_train,batch_size = 32, epochs =100)
````

````python
Y_pred = ann.predict(X_test)
print(Y_pred)
````

````python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(Y_pred,Y_test))
````

# ANN FOR CONCRETE COMPRESSIVE STRENGTH DATASET


```
# LINK FOR DATASET: https://archive.ics.uci.edu/dataset/165/concrete+compressive+strength
```



````python
data = pd.read_excel('/content/Concrete_Data.xls')
````

````python
data.head()
````

````python
new_column_names = ['col' + str(i+1) for i in range(len(data.columns) - 1)] + ['target_col']
data = data.rename(columns=dict(zip(data.columns, new_column_names)))
````

````python
data.head()
````

````python
x = data.iloc[:, :-1].values
y = data.iloc[:,-1].values
````

````python
print('X = ',x)
print('Y = ',y)
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
````

````python
ann = tf.keras.models.Sequential()
````

````python
ann.add(tf.keras.layers.Dense(units = 6,activation = 'elu'))
ann.add(tf.keras.layers.Dense(units = 6,activation = 'elu'))
ann.add(tf.keras.layers.Dense(units = 1))
````

````python
ann.compile(optimizer='adam', loss = 'mean_squared_error')
````

````python
ann.fit(X_train,Y_train,batch_size= 64,epochs = 150)
````

````python
Y_pred = ann.predict(X_test)
````

````python
print(Y_pred)
````

````python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(Y_pred,Y_test))
````
